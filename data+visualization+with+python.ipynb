```python
x = [1, 2, 3]
y = [1, 4, 9]
plt.plot(x, y, marker = "v", color = "orange")
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.scatter(x, y)
plt.scatter(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
x = np.linspace(0,16,100)

y = x**2
z = np.sqrt(x)
```

```python
fig,ax1 = plt.subplots()
color1 = 'red'
ax1.plot(x,y, color=color1)
ax1.tick_params(axis='y',color=color1)

color2 = 'blue'
ax2 = ax1.twinx()
ax2.plot(x,z, color=color2)
ax2.tick_params(axis='y',color=color2)
ax2.spines['left'].set_color(color1)
ax1.spines['right'].set_color(color2)
```
